# サブクエリを使ってみよう

サブクエリ(副問合せ)とは文字通りクエリの中に書くクエリのことを言います。

まずは次のセルを実行してみましょう。

```python
%%sql
SELECT
  id,
  name,
  birthday,
  gender,
  ph.count_purchase
FROM users
INNER JOIN
(
  SELECT user_id, count(user_id) AS count_purchase
  FROM purchase_histories
  GROUP BY user_id
) AS ph
ON users.id = ph.user_id;
```


顧客データとその顧客の購入数が出力されていれば成功です。

実行したクエリの()の中にさらにクエリがあります。これがサブクエリです。

複雑ですが1つずつ順を追って考えてみましょう。

## 確認テスト

(1) studentsテーブルとgradesテーブルから学生のデータとその学生の全教科の平均評定を取り出して、その中で1番高い平均評定を入力してください。
  ※この問題はSQLを穴埋めの形になっています。足りない部分を埋めてSQLを実行しましょう。

```python
%%sql
SELECT
    students.id,
    students.name,
    students.grade_level,
    g.avg_score_grades
FROM students
INNER JOIN
(
 SELECT student_id,
     avg(score) AS avg_score_grades
     FROM grades
GROUP BY student_id
) AS g
ON students.id = g.student_id
ORDER BY avg_score_grades DESC
```
