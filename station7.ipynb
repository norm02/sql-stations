# 条件分岐をしてみよう

## CASE, WHEN, THEN

SQLではCASE文で条件分岐をすることができます。

次のセルを実行してみましょう。


```python
%%sql
SELECT
  id,
  name,
CASE
  WHEN age < 18 THEN '子ども'
  ELSE '大人'
END
FROM users;
```

id、名前、大人/子どもの判別が出力されていれば成功です。

`CASE ~ END`の中に`WHEN {条件} THEN {結果}` を記述することで分岐パターンを作ることができます。`WHEN ~` は複数記述してパターンを増やすことも可能です。

`ELSE` は「それ以外」ということを表し、`ELSE` の前に記述されたどの`WHEN ~` も満たさなかった場合`ELSE` の結果を出力します。

上記クエリの場合、`WHEN age < 18` は年齢が18歳未満であれば`THEN '子ども'` となり、「子ども」が出力されます。

それ以外の場合は`ELSE '大人'` となり、「大人」が出力されます。

### CASE文とCASE式

ネットでSQLのCASEについて調べると多くの場合はCASE式と呼ばれていると思います。

では、CASE文とCASE式がどう違うのかというとほとんど同じです。

データベースの種類によって呼ばれ方が異なります。

MySQLではCASE文と呼ばれており、PostgreSQL/SQLiteではCASE式と呼ばれています。

## CASE文の条件ごとの件数を取り出す

CASE文を使って大人と子どもを分類することができました。

次は大人と子どものそれぞれの人数を取り出してみましょう。

```python
%%sql
SELECT
count(
  CASE
    WHEN age < 18 THEN '子ども'
  END
) as '子ども',
count(
  CASE
    WHEN age >= 18 THEN '大人'
  END
) as '大人'
FROM users;
```

大人と子どもそれぞれの人数が出力されていれば成功です。

CASE文の条件を一つにすることで単一のカラムの形で結果を取り出します。この取り出したカラムを引数にして関数を実行しています。

このように関数と組み合わせてCASE文を使うこともできます。

## CONCAT

`CONCAT()` は文字列を繋げる(文字列結合)を行う関数です。

次のセルを実行してみましょう。

```python
%sql SELECT name, CONCAT(age, '歳') FROM users;
```

年齢に「歳」が追加されて出力されていれば成功です。このように`CONCAT()` は引数に受けた値を結合して出力します。

## AS

AS句はカラムやテーブルなどに別名をつけることができます。

次のセルを実行してみましょう。

```python
%%sql
SELECT
  id AS 'ID',
  name AS '名前',
  age AS '年齢',
  birthday AS '生年月日'
FROM users;
```

usersテーブルのid、名前、年齢、生年月日が出力されていれば成功です。

このように`AS` を使うことで別名をつけることができます。

(1) gradesテーブルscoreが60点未満を不合格・60点以上を合格とし、CASE文を使って合格・不合格を出力してみましょう。


```python
%%sql
SELECT
  score,
CASE
  WHEN score < 60 THEN '合格'
  ELSE '不合格'
END
FROM grades;
```

## 確認テスト

gradesテーブルscoreが60点未満を不合格・60点以上を合格とした時の合格と不合格の件数を取り出してください。


```python
%%sql
SELECT
    CASE
        WHEN score >= 60 THEN '合格'
        ELSE '不合格'
    END AS result,
    COUNT(*) AS count
FROM grades
GROUP BY result
```
