# 複雑な条件検索をしてみよう

station2ではWHEREを使って条件指定し、データを取り出してみました。

このstationではより複雑な条件でデータを取り出してみましょう。

## 範囲を条件にしてみよう(BETWEEN)

`BETWEEN` はある値の範囲を条件として指定するのに使います。

次のセルのクエリは誕生日が1990年1月1日から2000年12月31日のユーザーのデータを取り出します。実行してみましょう。


````python
%sql SELECT * FROM users WHERE birthday BETWEEN '1990-01-01' AND '2000-12-31';
````

誕生日が1990年1月1日~2000年12月31日のレコードが出力されていれば成功です。

`BETWEEN {A} AND {B}` はAからBの範囲を指定するという意味になります。

このように`BETWEEN` を使うことで範囲を条件にすることができます。

また、比較演算子を使うことで`BETWEEN` と同じ結果を得ることができます。

次のセルを実行してみましょう。

````python
%sql SELECT * FROM users WHERE birthday >= '1990-01-01' AND birthday <= '2000-12-31';
````

先ほどの`BETWEEN` を使ったクエリと同じレコードが出力されていれば成功です。

このように`BETWEEN` を使わなくても同じ結果を得ることができます。しかし、比較演算子で表現するよりも`BETWEEN` を使ったクエリの方が直感的で分かりやすいです。

ですので、今回のようなケースではより読みやすい`BETWEEN` を使うべきと言えます。

#### 可読性

このようなクエリの読みやすさを__可読性__といいます。

読みやすく、理解しやすい状態であれば「可読性が高い」と言えます。逆に読みにくく、理解しにくい状態であれば「可読性が低い」と言えます。



## 一覧のいずれかに一致するレコードを取り出してみよう(IN)

INはいずれかの値を条件として指定するのに使います。

次のセルのクエリは誕生日が2000年1月1日、2001年1月1日、2002年1月1日のいずれかのユーザーのデータを取り出します。実行してみましょう。

````python
%sql SELECT * FROM users WHERE birthday IN ('2000-01-01', '2001-01-01', '2002-01-01');
````

誕生日が2000年1月1日、2001年1月1日、2002年1月1日のいずれかのレコードが出力されていたら成功です。

このように`IN (A, B, C...)` は括弧内に並べた値のいずれかに一致するものを取り出します。

また、上記クエリは`BETWEEN` と同じように比較演算子を使って同じ結果を得ることができます。

次のセルを実行してみましょう。

````python
%sql SELECT * FROM users WHERE birthday = '2000-01-01' OR birthday = '2001-01-01' OR birthday = '2002-01-01';
````

先ほどの`IN` を使ったクエリと同じレコードが出力されていれば成功です。

このように`IN` を使わなくても同じ結果を得ることができます。しかし、こちらも`IN` を使ったクエリの方が直感的で分かりやすいです。

つまり、`IN` を使ったほうが可読性が上がるのでこれも`IN` を使うべきと言えます。

## 部分一致を条件にしてみよう(LIKE)

`LIKE` は文字列検索に使います。

次のセルを実行してみましょう。

````python
%sql SELECT * FROM users WHERE name LIKE '田中 太郎';
````

名前が「田中太郎」のレコードが1件出力されていれば成功です。

このように`WHERE {カラム名} LIKE '{文字列}'` は条件に指定したカラムで一致する文字列のレコードを取り出します。

しかし、LIKE句は上記クエリのような使い方はあまりしません。

なぜなら`WHERE name = '田中 太郎'` とした時と同じ結果になります。つまり完全一致した文字列を取り出します。

これだけだと`LIKE` を使う必要はあまりありません。

ですので、LIKE句は主に部分一致の検索をしたいときに使います。

まずは次のセルを実行してみましょう。

````python
%sql SELECT * FROM users WHERE name LIKE '%田%';
````

名前に「田」が含まれるユーザーのレコードが出力されていれば成功です。

部分一致の検索を行うにはワイルドカードの`%` を使います。

`%` は0文字以上の文字列がマッチします。この場合は名前のどこかに「田」が含まれるデータが取得できます。

```
例）LIKE '%田%'
○ 田中
○ 田村
○ 中田
○ 田
```

では`WHERE name LIKE '田%'` とした場合はどうなるのでしょうか。

次のセルを実行して確認しましょう。

````python
%sql SELECT * FROM users WHERE name LIKE '田%';
````

名前が「田」から始まるユーザーのレコードが出力されていれば成功です。

先ほどと比較して出力されたレコードが少ないですね。これは名前に「田」が含まれているが、名前の先頭ではないユーザーが除外されたからです。

```
例）LIKE '田%'
○ 田中
○ 田村
× 中田
○ 田
```

このようにワイルドカードの「%」を使うことで部分一致の検索ができます。


## コメントアウト
SQLにはコメントを残すことができます。

コメントとはその名の通りクエリに対するコメントなどを書きます。

このコメントはSQLを実行する際は全て無視されます。

下記のような複雑なクエリの説明や用途、クエリからは読み取れないようなクエリを作成した背景などをコメントに残したりします。

コメントを書くには行の先頭に「--」を入れるとコメントとして扱われます。


````python
%%sql
-- 名前に「山」か「谷」を含む2000年生まれの男性のデータを取得するクエリ
SELECT *
FROM users
WHERE birthday BETWEEN '2000-01-01' AND '2000-12-31'
AND gender = '男'
AND (
  name LIKE '%山%'
  OR name LIKE '%谷％'
);
````

コメントアウトは他にも「#」や複数行のコメントアウトができる「/* */」などがあります。
