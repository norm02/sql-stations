## 複数テーブルで管理しているデータをまとめて取り出す

ではテーブルを分けて管理されている顧客データとそれに紐づく購入履歴を取り出すには、テーブル同士を結合させることで実現できます。

purchase_historiesテーブルには顧客データと紐付けるためのuser_idを保持しています。このuser_idと同じ値を持つusersテーブルのidのレコードを繋げることで結合できます。

では、次のセルを実行してみましょう。


```python
%%sql
SELECT *
FROM users
INNER JOIN purchase_histories
ON users.id = purchase_histories.user_id;
```

usersテーブルとpurchase_historiesテーブルを結合したものが出力されていれば成功です。

では、SQLの中身について説明します。

`INNSER JOIN {結合するテーブル名}` で結合するテーブルを指定します。

`ON {結合するそれぞれのカラム}` で結合することができます。

![station6_04](https://drive.google.com/uc?id=1GS_mZPsx6AxcFN3q6ARaRvRhhfasv_u0)

## 内部結合と外部結合

テーブル結合には大きく分けて内部結合と外部結合があります。

`INNER JOIN` は内部結合です。

内部結合は結合に成功したレコードのみを取り出します。

つまり、結合できなかったレコードは取り出しません。

usersテーブルにはAさんとBさんのデータが入っています。

Aさんはノートパソコンとマウスを購入しています。しかし、Bさんは何も購入していないので購入履歴がありません。

なので、先ほどの`INNER JOIN` を使ったクエリでBさんのデータを取り出すことはできませんでした。

結合できないレコードも全て取り出したい時は外部結合を使います。

```python
%%sql
SELECT *
FROM users
LEFT JOIN purchase_histories
ON users.id = purchase_histories.user_id;
```

usersテーブルとpurchase_historiesテーブルを結合したものが出力されていれば成功です。

外部結合には`LEFT JOIN` の左外部結合、`RIGHT JOIN`の右外部結合の2種類があります。

左外部結合と右外部結合の違いは基準にするテーブルを左側にするか右側にするかの違いです。

左外部結合の場合は左のテーブルのレコードを全て取り出し、対応する右のテーブルのレコードがない場合はNULLとして取り出されます。

※GoogleColaboratory(Jupyter Notebook)上で実行している都合上`NULL` が`None` と出力されていますが、`None` は`NULL` を指しています。

### NULL

NULLとは何もないということを意味し、データがないことを示します。

つまり、外部結合などで取り出してNULLと表示されたものは値が入っていないことを意味します。

## DISTINCT

`DISTINCT` は取り出すデータの重複を排除します。

次のセルを実行してみましょう。購入履歴テーブルから購入履歴のあるユニークなユーザーを取り出すクエリです。

```python
%sql SELECT DISTINCT user_id FROM purchase_histories;
```

user_idが重複せずに出力されていれば成功です。

このような重複を排除して取り出したいケースに使うことができます。

ただし、クエリのパフォーマンスの観点から`DISTINCT` を使わず、重複せずに取り出すことが可能なら使うべきでないです。

なぜかというと`DISTINCT` はDBから取り出した後のデータの重複を排除します。最初から重複せず取り出すことが可能ならその方が効率的といえます。

DBの性能にもよりますが、データが数十件・数百件ほどであればそれほどの差はありません。しかし、数十万件・数百万件などの膨大なデータ量になればクエリ効率が顕著に現れます。

クエリの効率を考えてクエリを作れるようになるとより良いので意識していきましょう。
