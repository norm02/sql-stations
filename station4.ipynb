# グルーピング/並び替えをしてみよう

このstationでは取得したデータのグルーピングと並び替えをしてみましょう。

## 取り出したレコードを並び替えてみよう(ORDER BY)

SQLで取得するデータをある特定の並び順にして取得したいときは`ORDER BY` を使うことで並び順を指定することができます。

次のセルを実行してみましょう。

```python
%sql SELECT * FROM users ORDER BY age ASC;
```

年齢を昇順でユーザーのレコードが全件出力されていれば成功です。

`ORDER BY age ASC` はageカラムの値を使って昇順に並び替えます。

構文としては`ORDER BY {カラム} {ASC or DESC}` となります。

`DESC` が降順、`ASC` が昇順を意味します。DESC/ASCの指定がない場合はASC(昇順)となります。


## グループ化してみよう(GROUP BY)

例えば、性別ごとの人数と取り出すなどの特定カテゴリの数を取り出すにはどのようにすればいいのでしょうか。

次のセルを実行してみましょう。性別ごとの人数を取り出すクエリです。

```python
%sql SELECT gender, count(gender) FROM users GROUP BY gender;
```

性別ごとの人数が出力されていれば成功です。

データをグループ化するには`GROUP BY {カラム名}` を使って、グループ化することができます。

つまり、`GROUP BY gender` は性別のカラムの値でグループ化するという意味になります。

`SELECT count(gender)` は集計関数というものを使ってグループ化したものの合計値を取り出しています。

詳しくは次のstationで扱いますが、今は合計値を取り出していることが理解できれば問題ありません。
